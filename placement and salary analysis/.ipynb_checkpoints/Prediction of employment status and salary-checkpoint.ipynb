To select the best features we will now check the scores of the dependency of y1 on X1

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
y1=pd.DataFrame(y1)
#apply SelectKBest class to extract top 14 best features
bestfeatures = SelectKBest(score_func=chi2, k=14)
fit = bestfeatures.fit(X1,y1)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X1.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(14,'Score'))  #print 16 best features
```

Even though some of these have low scores we cannot drop these columns as these can be crucial data.

To select the best features we will now check the scores of the dependency of y2 on X2

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
y2=pd.DataFrame(y2)
#apply SelectKBest class to extract top 15 best features
bestfeatures = SelectKBest(score_func=chi2, k=15)
fit = bestfeatures.fit(X2,y2)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X2.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(15,'Score'))  #print 17 best features
```

## Train test split of data

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size = 0.2, random_state = 42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size = 0.2, random_state = 42)
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train1=sc_X.fit_transform(X_train1)
X_test1=sc_X.transform(X_test1)
```

```python
sc_X2 = StandardScaler()
X_train2 = sc_X2.fit_transform(X_train2)
X_test2 = sc_X2.transform(X_test2)
sc_y = StandardScaler()
y_train2 = sc_y.fit_transform(y_train2)
```

# Status prediction(placed/not placed)
We will use different models to predict the status:
* Logistic Regression.<br/>
* Kernel SVM.<br/>
* Random Forest Classification.

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(random_state = 42)
log.fit(X_train1,y_train1)
#Predicting test set results
y_pred_log = log.predict(X_test1)
#making the confusion matrix
from sklearn.metrics import confusion_matrix
cm_log = confusion_matrix(y_test1, y_pred_log)
print(cm_log)
```

```python
#Kernel SVM classification
from sklearn.svm import SVC
svm = SVC(kernel='rbf', random_state=42)
svm.fit(X_train1,y_train1)
# Predicting the Test set results
y_pred_svm = svm.predict(X_test1)
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm_svm = confusion_matrix(y_test1, y_pred_svm)
print(cm_svm)
```

```python
#Random Forest classification
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(criterion = 'entropy', n_estimators=100, random_state=42)
rfc.fit(X_train1, y_train1)
# Predicting the Test set results
y_pred_rfc = rfc.predict(X_test1)
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm_rfc = confusion_matrix(y_test1, y_pred_rfc)
print(cm_rfc)
```

Removing some features. we can see that our model is giving us some preety good results but we can improve the accuracy by dropping a column or two. I tried dropping the 'ssc_b' column and 'hsc_b' column and the accuracy of my model increased drastically.

```python
x1=100*(cm_log[0][0]+cm_log[1][1])/np.sum(cm_log)
```

```python
print("Test Accuracy of Logistic regression with 80:20 train test split is ",x1,"%")
```

# Salary prediction
We will use different models to predict the status:

* Multiple Linear Regression.<br/>
* Support Vector Regression(SVR).<br/>
* Random Forest Regression.

```python
#Multiple linear regression
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(X_train2, y_train2)
# Predicting the Test set results
y_pred_mlr = sc_y.inverse_transform(mlr.predict(X_test2))
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='poly')
svr.fit(X_train2, y_train2)

# Predicting a new result
y_pred_svr = sc_y.inverse_transform(svr.predict(X_test2))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators = 300, random_state = 42)
rfr.fit(X_train2, y_train2)

# Predicting a new result
y_pred_rfr = sc_y.inverse_transform(rfr.predict(X_test2))
```

```python
a = pd.DataFrame({'Actual Salary':list(y_test2.values),'MLR_Salary':list(y_pred_mlr),'SVR_Salary':list(y_pred_svr),'RFR_Salary':list(y_pred_rfr)})
a
```

We can see that the accuracy is not varying much using different models as there are people who aren't placed and hence their salary is 0 and their predicted salary will be low but not zero hence the accuracy will not be very variant.
